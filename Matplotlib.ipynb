# Matplotlib


**Matplotlib** is a popular Python library used for creating static, interactive, and animated visualizations. It allows users to plot data in various formats, such as line graphs, bar charts, scatter plots, and more, making it easier to understand and analyze data visually.
 
 
 
**Seaborn** is built on top of Matplotlib and provides a higher-level interface for creating more attractive and informative statistical graphics. It simplifies the process of making complex visualizations and comes with built-in themes and color palettes to enhance the appearance of your plots. Seaborn is particularly good for visualizing relationships in data and handling statistical representations.

```python
import matplotlib.pyplot as plt
```

# Line Plot

```python
x =[1,2,3,4,5]
y =[2,3,5,7,11]

plt.plot(x,y)
plt.title("Basic Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

# Multi Line Plots

```python
y2 = [1,2,4,8,16]

plt.plot(x,y, label = "y = primes")
plt.plot(x,y2, label = "y2 = powers of 2")

plt.title("Line graph with multiple lines")
plt.legend()

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# Bar Chart

```python
categories = ["A","B","C","D"]

values = [4,7,1,8]

plt.bar(categories, values)

plt.title("Bar Chart")

plt.xlabel("Categories")

plt.ylabel("Values")

plt.show()
```

### HORIZONTAL PLOT

```python
plt.barh(categories, values)

plt.title("Bar Chart")

plt.xlabel("Categories")

plt.ylabel("Values")

plt.show()
```

### Stacked Bar Chart

```python
values2 = [3,5,2,6]
plt.bar(categories, values, label="Dataset")
plt.bar(categories, values2,bottom=values, label="Dataset2")
plt.legend()

```

# Histograms

```python
import numpy as np
data = np.random.randn(1000)
data

plt.hist(data,bins=30, alpha=0.7, color= "red", edgecolor= "black")
```

# Scatter Plot

```python
sizes = 100 * np.random.rand(50)
sizes
colors = np.random.rand(50)
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y, s=sizes, c=colors)
```

# Pie Chat

```python
sizes = [15,30,45,10]
labels = ["A","B","C","D"]

plt.pie(sizes,labels= labels)

plt.title("Pie Chat")

plt.show()
```

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
# Load a sample dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Scatter plot")
plt.show()
```

```python
sns.histplot(tips["total_bill"], bins=30, kde=True)
```

```python
sns.violinplot(data=tips, x="day", y="total_bill")
```

```python
sns.pairplot(data=tips, hue="sex")
```

```python
tips2=tips[["total_bill","tip","size"]]
coro = tips2.corr()
```

```python
sns.heatmap(coro,annot=True)
plt.savefig("Heatmap.png")
```
